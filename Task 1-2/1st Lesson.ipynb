```python
def highlight_face_parts(image):
    cv2.circle(img=image, center=(2000, 1880), radius=200, color=(0, 255, 0), thickness=50)
    cv2.circle(img=image, center=(3050, 1880), radius=200, color=(0, 0, 255), thickness=50)
    
    nose_vertices = np.array([[2500, 2750], [2300, 2500], [2700, 2500]], np.int32)
    cv2.polylines(image, [nose_vertices], isClosed=True, color=(255, 0, 0), thickness=50)
    # cv2.fillPoly(image, [nose_vertices], color=(0, 0, 255))
    
    
    left_ear_vertices = nose_vertices = np.array([[1400, 1230], [1320, 888], [1290, 330],
                                                  [1360, 280], [1640, 390], [1960, 660]], np.int32)
    
    right_ear_vertices = nose_vertices = np.array([[2500, 2750], [2300, 2500], [2700, 2500]], np.int32)

    
    cv2.polylines(image, [left_ear_vertices], isClosed=True, color=(0, 0, 255), thickness=50)
    cv2.polylines(image, [right_ear_vertices], isClosed=True, color=(0, 0, 255), thickness=50)
    # cv2.fillPoly(image, [nose_vertices], color=(0, 0, 255))
    return image
    
```

```python
cat_face = np.array(Image.open(r"cat_face.jpg"))

plt.figure(figsize=(16, 9), dpi=100)
plt.subplot(1, 2, 1)
plt.imshow(cat_face)
plt.subplot(1, 2, 2)
highlight_face_parts(cat_face)
plt.imshow(cat_face)
```
